# Task 3

```python
def plot_lines(n) -> None:
    '''
    Plot lines with different styles.

    Parameters
    ----------
    n : the number of lines
    '''

    points = random.randint(5, 10)
    plt.figure(figsize = (8, 6))
    plt.title(f'Plotting {n} lines')
    colors = ['green', 'red', 'yellow', 'magenta', 'black', 'orange']

    for i in range(n):
        plt.plot([x for x in range(points)], [x ** i for x in range(points)], label = f'x^{i}', linewidth = random.randint(5, 15), color = random.choice(colors))
        plt.legend(bbox_to_anchor = (1.05, 1), loc = 'upper left')

    plt.show()


plot_lines(4)
```

# Task 7

```python
x = ['Python', 'Java', 'C', 'C++', 'R', 'JavaScript', 'C#']
popularity = [100, 96.3, 94.4, 87.5, 81.5, 79.4, 74.5]
x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, popularity, color = 'blue')
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Language\n' + 'IEEE Survey 2019')
plt.xticks(x_pos, x)
plt.minorticks_on()
plt.grid(which = 'major', linestyle = '-', linewidth = '0.5', color = 'red')
plt.grid(which = 'minor', linestyle = ':', linewidth = '0.5', color = 'black')
plt.show()
```
